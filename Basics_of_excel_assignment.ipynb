1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.

 -> In Excel (or any spreadsheet software), cell referencing determines how a formula behaves when it is copied or moved to another cell. There are three main types of cell references: Relative, Absolute, and Mixed.


1. Relative Cell Reference

Definition: Changes automatically when a formula is copied to another cell.

Symbol: No dollar sign ($).

Example: =A1 + B1

➡️ If you copy this formula from row 1 to row 2, it becomes =A2 + B2.

Excel adjusts the reference relative to its new position.

Use Case: When you want the formula to apply to multiple rows or columns automatically.

2. Absolute Cell Reference

Definition: The reference remains fixed no matter where the formula is copied.

Symbol: Both column and row are locked using $.

Example: =$A$1 + B1

➡️ If you copy this formula anywhere, $A$1 will always point to cell A1, while B1 will adjust.

Use Case: When referring to a constant value or fixed cell (e.g., tax rate, conversion rate).

3. Mixed Cell Reference

Definition: Only one part (either the row or column) is locked.

Symbol: One dollar sign ($) is used.

Examples:

=$A1 → Column A is fixed, row changes.

=A$1 → Row 1 is fixed, column changes.

➡️ If you copy =$A1 down, row changes but the column stays as A.
➡️ If you copy =A$1 across, column changes but row stays as 1.

Use Case: When you want partial flexibility — such as when building multiplication tables or structured references.

Summary Table


Type of Reference	Example	Behavior When Copied	Typical Use
Relative	A1	Both row & column change	Repetitive calculations across cells
Absolute	$A$1	Both row & column fixed	Fixed values like constants
Mixed	$A1 or A$1	Only one part fixed	Partial locking (row or column)
